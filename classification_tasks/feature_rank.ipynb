```python
%load_ext autoreload
%autoreload 2
import os
import glob
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick
import seaborn as sns
%matplotlib inline
sns.set(font_scale=1.1)

np.random.seed(1)
np.set_printoptions(precision=6, suppress=True)
plot_height, legend_size = 10, 18
marker_size, line_width = 2.0, 1.2
label_size = 16
dashed_linestyle = (0, (10, 10))  # 'dashed'
```

```python
### Final plots

colors = ['lightsteelblue', 'cornflowerblue', 'blue', 'midnightblue', 'black', 'green', 'orange', 'red']
label_size = 16.3
figsize = (6.5, 4.7)  # (7.0, 4.4)

model, model_width = 'resnet18', 64        
lr, label_smoothing = '', 0.0  

for percentile_pcs in [0.99]:
    for dataset in ['cifar10', 'cifar100', 'tiny_imagenet']:
        n_train = 100000 if dataset == 'tiny_imagenet' else 50000
        for exp_name, lr_schedule in zip(['sam_low_rank_basic', 'sam_low_rank_augm_sota'], ['constant', 'piecewise_05epochs']):
            l2_reg = '0.0005' if (dataset, exp_name) == ('tiny_imagenet', 'sam_low_rank_augm_sota') else ''  # also runs with 5e-05 are available but the delta is smaller there
            glob_str = 'exps/{}/*dataset={} model={}*lr_max={}*l2_reg={}*lr_schedule={}*'.format(exp_name, dataset, model, lr, l2_reg, lr_schedule)
            print(glob_str)
            model_files = glob.glob(glob_str)
            print(len(model_files))

            if (dataset, exp_name) == ('cifar10', 'sam_low_rank_basic'):
                rhos = [0.0, 0.1, 0.2, 0.4, 0.6]
            elif (dataset, exp_name) == ('cifar10', 'sam_low_rank_augm_sota'):
                rhos = [0.0, 0.1, 0.4, 0.6, 0.8]
            elif (dataset, exp_name) == ('cifar100', 'sam_low_rank_basic'):
                rhos = [0.0, 0.1, 0.2, 0.4, 0.6]
            elif (dataset, exp_name) == ('cifar100', 'sam_low_rank_augm_sota'):
                rhos = [0.0, 0.2, 0.6, 0.9, 1.2]
            elif (dataset, exp_name) == ('tiny_imagenet', 'sam_low_rank_basic'):
                rhos = [0.0, 0.1, 0.4, 0.5, 0.6]
            elif (dataset, exp_name) == ('tiny_imagenet', 'sam_low_rank_augm_sota'):
                rhos = [0.0, 0.1, 0.4, 0.6, 0.7]
            else:
                rhos = 'all'
            model_files_filtered = []
            for model_file in model_files:
                metrics = np.load(model_file, allow_pickle=True).item()
                print(metrics['label_smoothing'])
                if metrics['test_err'][-1] <= 0.99 and metrics['label_smoothing'] == label_smoothing and \
                (rhos == 'all' or metrics['sam_rho'] in rhos):
                    model_files_filtered.append(model_file)
            model_files = model_files_filtered

            model_files = sorted(model_files, key=lambda s: s.split('sam_rho')[1])
            print(model_files)
            
            plt.figure(figsize=figsize)
            for i in range(len(model_files)):
                metrics = np.load(model_files[i], allow_pickle=True).item()
                print(metrics['train_loss'][-1], model_files[i])
                label = '$\\rho$={}'.format(metrics['sam_rho'])
                epochs = np.array(metrics['iter']) / (n_train / metrics['batch_size'])
                ax = sns.lineplot(x=epochs, y=metrics['train_loss'], linewidth=line_width, markersize=marker_size, label=label, color=colors[i])
                # ax = sns.lineplot(x=epochs, y=metrics['train_loss_swa'], linewidth=line_width, markersize=marker_size, label=label, color=colors[i], linestyle='--')
            ax.set_yscale('log')
            ax.set_xlabel('Epoch', size=label_size)
            ax.set_ylabel('Training loss', size=label_size)
            ax.legend(loc='best')
            plt.savefig('plots/sam_low_rank-train_loss-dataset={}_model={}_exp_name={}_lr={}.pdf'.format(dataset, model, exp_name, lr), bbox_inches='tight')
            plt.show()


            plt.figure(figsize=figsize)
            for i in range(len(model_files)):
                metrics = np.load(model_files[i], allow_pickle=True).item()
                print(metrics['test_err'][-1], min(metrics['test_err_swa']), model_files[i])
                label = '$\\rho$={}'.format(metrics['sam_rho'])
                epochs = np.array(metrics['iter']) / (n_train / metrics['batch_size'])
                ax = sns.lineplot(x=epochs, y=metrics['test_err'], linewidth=line_width, markersize=marker_size, label=label, color=colors[i])
                # ax = sns.lineplot(x=epochs, y=metrics['test_err_swa'], linewidth=line_width, markersize=marker_size, label=label, color=colors[i], linestyle='--')
            if (dataset, exp_name) == ('cifar10', 'sam_low_rank_basic'):
                ax.set_ylim(0.07, 0.3)
            elif (dataset, exp_name) == ('cifar10', 'sam_low_rank_augm_sota'):
                ax.set_ylim(0.033, 0.25)
            elif (dataset, exp_name) == ('cifar100', 'sam_low_rank_basic'):
                ax.set_ylim(0.275, 0.6)
            elif (dataset, exp_name) == ('cifar100', 'sam_low_rank_augm_sota'):
                ax.set_ylim(0.19, 0.5)
            elif (dataset, exp_name) == ('tiny_imagenet', 'sam_low_rank_basic'):
                ax.set_ylim(0.38, 0.7)
            elif (dataset, exp_name) == ('tiny_imagenet', 'sam_low_rank_augm_sota'):
                ax.set_ylim(0.3, 0.68)
            ax.set_xlabel('Epoch', size=label_size)
            ax.set_ylabel('Test error', size=label_size)
            ax.set_yticklabels(['{:.0%}'.format(x) for x in ax.get_yticks()])
            ax.legend(loc='best')
            plt.savefig('plots/sam_low_rank-test_err-dataset={}_model={}_exp_name={}_lr={}.pdf'.format(dataset, model, exp_name, lr), bbox_inches='tight')
            plt.show()


            plt.figure(figsize=figsize)
            i_layer = 0
            for i in range(len(model_files)):
                metrics = np.load(model_files[i], allow_pickle=True).item()
                label = '$\\rho$={}'.format(metrics['sam_rho'])
                epochs = np.array(metrics['iter']) / (n_train / metrics['batch_size'])
                feature_ranks = [np.sum(np.cumsum(svals[i_layer]**2) <= np.sum(svals[i_layer]**2) * percentile_pcs) + 1 for svals in metrics['feature_sing_vals']]
                print(feature_ranks[-1], model_files[i])
                ax = sns.lineplot(x=epochs, y=feature_ranks, linewidth=line_width, markersize=marker_size, label=label, color=colors[i])
            ax.set_xlabel('Epoch', size=label_size)
            ax.set_ylabel('Feature rank (interm. block)', size=label_size)
            ax.legend(loc='best')
            plt.savefig('plots/sam_low_rank-feature_rank-i_layer={}-dataset={}_model={}_exp_name={}_lr={}_percentile={}.pdf'.format(i_layer, dataset, model, exp_name, lr, percentile_pcs), bbox_inches='tight')
            plt.show()
            

            plt.figure(figsize=figsize)
            i_layer = 0
            for i in range(len(model_files)):
                metrics = np.load(model_files[i], allow_pickle=True).item()
                label = '$\\rho$={}'.format(metrics['sam_rho'])
                epochs = np.array(metrics['iter']) / (n_train / metrics['batch_size'])
                knn_errors = [errs[i_layer] for errs in metrics['test_knn_errors']]
                print(knn_errors[-1], model_files[i])
                ax = sns.lineplot(x=epochs, y=knn_errors, linewidth=line_width, markersize=marker_size, label=label, color=colors[i])
            if (dataset, exp_name) == ('cifar10', 'sam_low_rank_basic'):
                ax.set_ylim(0.615, 0.85)
            elif (dataset, exp_name) == ('cifar10', 'sam_low_rank_augm_sota'):
                ax.set_ylim(0.595, 0.85)
            elif (dataset, exp_name) == ('cifar100', 'sam_low_rank_basic'):
                ax.set_ylim(0.24, 0.45)
            elif (dataset, exp_name) == ('cifar100', 'sam_low_rank_augm_sota'):
                ax.set_ylim(0.23, 0.37)
            elif (dataset, exp_name) == ('tiny_imagenet', 'sam_low_rank_basic'):
                ax.set_ylim(0.28, 0.5)
            elif (dataset, exp_name) == ('tiny_imagenet', 'sam_low_rank_augm_sota'):
                ax.set_ylim(0.27, 0.5)
            ax.set_xlabel('Epoch', size=label_size)
            ax.set_ylabel('kNN error on {} (interm. block)'.format(
                'C100' if dataset == 'cifar10' else 'C10'), size=label_size)
            ax.set_yticklabels(['{:.0%}'.format(x) for x in ax.get_yticks()])
            ax.legend(loc='best')
            plt.savefig('plots/sam_low_rank-knn_err-i_layer={}-dataset={}_model={}_exp_name={}_lr={}.pdf'.format(i_layer, dataset, model, exp_name, lr), bbox_inches='tight')
            plt.show()
```
